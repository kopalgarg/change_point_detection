```python
def GaussianKernelFunc(x,y, sigma=1):
    if (sigma is None):
        sigma = 1
    nSamp = np.shape(x)[0]
    if nSamp == 1:
        return np.exp(-(x-y)**2/(2*sigma))

    if (np.ndim(x)==1):
        x=np.expand_dims(x,0)
    if (np.ndim(y)==1):
        y=np.expand_dims(y,0)
    M = ot.dist(x,y,metric = 'sqeuclidean')
    return np.exp(-M/(2*sigma))

def MmdGaussianKernelFunc(x,y=None,sigma=1, diag=0):
    if (sigma is None):
        sigma = 1
    if (y is None):
        y=x
    dist = GaussianKernelFunc(x,y,sigma=sigma)
    if (diag == 0 ):
        return np.sum(dist)-np.sum(np.diag(dist)) # if we're looking just at x-x, remove diagonal
    else:
        return np.sum(dist) # if we're looking just at x-x, remove diagonal

def ComputeMmdDistance(cloud_i,cloud_j, sigma = None):
    m = cloud_i.shape[0]
    n = cloud_j.shape[0]

    if (sigma is None):
        # Comput the median trick
        fullDat = np.append(cloud_i,cloud_j, axis=0)
        distance = ot.utils.dist(fullDat, fullDat, metric = 'sqeuclidean')
        sig = np.median(distance.flatten())
    
    if (m==n):
        mmd = 1/(m*(m-1))*MmdGaussianKernelFunc(cloud_i, sigma=sigma)+ 1/(n*(n-1))*MmdGaussianKernelFunc(cloud_j, sigma=sigma)-2/(m*(m-1))*MmdGaussianKernelFunc(cloud_i,cloud_j, sigma=sigma)
        import pdb; pdb.set_trace()
    else:
        mmd = 1/(m*(m-1))*MmdGaussianKernelFunc(cloud_i, sigma=sigma)+ 1/(n*(n-1))*MmdGaussianKernelFunc(cloud_j, sigma=sigma)-2/(m*n)*MmdGaussianKernelFunc(cloud_i,cloud_j, sigma=sigma, diag=1)
    return mmd # Return square value. 


def ComputeOtDistance(cloud_i,cloud_j, metric = 'sqeuclidean'):
    nSamp_i = cloud_i.shape[0]
    nSamp_j = cloud_j.shape[0]
    a = np.ones(nSamp_i)/nSamp_i
    b = np.ones(nSamp_j)/nSamp_j
    M = ot.utils.dist(cloud_i,cloud_j, metric = metric)
    G = ot.emd(a,b,M)
    return np.sqrt(np.sum(np.multiply(G,M)))

def ComputeSinkhornDistance_variableLambd(cloud_i, cloud_j, metric = 'sqeuclidean'):
    lambdStart = 1e-10
    nSamp = cloud_i.shape[0]
    a = np.ones(nSamp)/nSamp

    M = ot.utils.dist(cloud_i, cloud_j, metric = metric)
    
    converged = 0
    lambd = lambdStart
    while (converged == 0):
        G = ot.sinkhorn(a,a,M,lambd)
        if (np.sum(G)==0):
            lambd = lambd*10
        else:
            converged = 1
    return np.sum(np.multiply(G,M))

```

```python
def TwoSampleWTest(sampA, sampB, step = None):
    lenA = len(sampA)
    lenB = len(sampB)
    (cdfX, cdfY) = F_GInv(sampA,sampB)
            
    if (step is None):
        distOut = DistanceSquaredToUniform(cdfX, cdfY)*(lenA*lenB)/(lenA+lenB)
    else:
        distOut = DistanceSquaredToUniform(cdfX, cdfY, step=step)*(lenA*lenB)/(lenA+lenB)
    return distOut
def DistanceSquaredToUniform(pp,cdf,step= 0.01):
    pp=np.append(pp,1)
    cdf = np.append(cdf,1)
    xAll = np.linspace(0,1,int(1/step)+1)
    total = 0 
    for x in xAll:
        argX = np.argwhere(pp>=x)
        total += (x - cdf[argX[0]])*(x - cdf[argX[0]])*step
    return total
def F_GInv(F,G):
    F=F.flatten()
    G=G.flatten()
    n = len(F)
    m = len(G)
    Fsort = np.sort(F)
    Gsort = np.sort(G)
    outX = np.zeros(m)
    outY = np.zeros(m)
    for i in range(m):
        dist = np.argwhere(Fsort <= Gsort[i])
        outY[i] = len(dist)/n
        outX[i] = (i+1)/m # cdf jumps at points in Gm
    return (outX, outY)

def Compute2SampleWStat(dat, window, stride):
    lenDat = len(dat)
    dim=len(dat[0])
    out = np.zeros((int(np.floor(lenDat/stride)), dim))
    count = 0
    for i in range(0,lenDat-stride,stride):
        for j in range(dim):
            if (i<window or i >= lenDat-window):
                out[count,j]=0
            else:        
                win1 = dat[i-window:i,j]
                win2 = dat[i:i+window,j]
                out[count,j] = TwoSampleWTest(win1, win2)
        count = count+1
    outSingle = np.mean(out,axis=1)
    return out
```

```python
prev_window_len = 10
next_window_len = 10
mean_diff = []
var_diff = []
W = []
for i in range(prev_window_len, series.shape[1]-next_window_len):
  prev_seg = series[:, i-prev_window_len:i]
  next_seg = series[:, i:i+next_window_len]
  mean_diff.append(np.average(prev_seg, axis=1)-np.average(next_seg, axis=1)) 
  var_diff.append(np.var(prev_seg, axis = 1)-np.var(next_seg, axis = 1))
  W.append(TwoSampleWTest(prev_seg, next_seg))


plt.plot(np.concatenate((np.zeros((prev_window_len, series.shape[0])),np.array(var_diff), np.zeros((next_window_len, series.shape[0])))))
plt.show()
plt.plot(np.concatenate((np.zeros((prev_window_len, series.shape[0])),np.array(mean_diff), np.zeros((next_window_len, series.shape[0])))))
plt.show()
plt.plot(np.transpose(series))
plt.show()
plt.plot(np.concatenate((np.zeros((prev_window_len)),np.array(W)[:,0], np.zeros((next_window_len)))))
np.array(W).shape
```

```python
prev_window_len = 10
next_window_len = 10
corr_ = []
det = []
trace_ = []
mean_diff = []
var_diff = []
W = []
# https://online.stat.psu.edu/stat505/lesson/1/1.5
series=np.concatenate((x2, x4,x1,x3))
for i in range(prev_window_len, series.shape[0]-next_window_len):
    prev_seg = series[i-prev_window_len:i]
    next_seg = series[i:i+next_window_len]
    corr_.append(pd.DataFrame(np.concatenate((prev_seg, next_seg))).corr()[0])
    det.append(np.linalg.det(np.cov(prev_seg, next_seg)))
    trace_.append(np.trace(np.cov(prev_seg, next_seg)))
    mean_diff.append(np.average(prev_seg, axis=1)-np.average(next_seg, axis=1)) 
    var_diff.append(np.var(prev_seg, axis = 1)-np.var(next_seg, axis = 1))
    W.append(TwoSampleWTest(prev_seg, next_seg))
```
